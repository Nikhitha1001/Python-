# Exploratory Data Analysis (EDA)

## Understanding realationships between:




###  Average Daily Usage (in hours) of social media by age

```python
# Group data by Age and calculate mean usage
age_usage = dataset.groupby("Age")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=age_usage, x="Age", y="Avg_Daily_Usage_Hours", marker='o', linewidth=2)

# Chart details
plt.title("Average Daily Social Media Usage by Age")
plt.xlabel("Age")
plt.ylabel("Average Daily Usage (Hours)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 In this vizualization we can see how adults are addicted by social media.
 18 age group student use the most and as 23 age people are of less usage.

Students aged 19 to 21 show the highest average daily usage of social media. Usage tends to slightly drop after early twenties.

This could indicate that younger students are more prone to excessive usage, possibly due to more free time or social peer pressure.




###  Average Daily Usage (in hours) of social media by Gender

```python
# Group data by Gender and calculate mean usage
age_usage = dataset.groupby("Gender")["Avg_Daily_Usage_Hours"].mean().reset_index()

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=age_usage, x="Gender", y="Avg_Daily_Usage_Hours", marker='o', linewidth=2)

# Chart details
plt.title("Average Daily Social Media Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Usage (Hours)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

On average, female students tend to spend more time on social media daily compared to male students, indicating a possible need for more targeted awareness among female users.

### Sleep patterns (Line chart)

```python
# Average Sleep Hours vs Age
sleep_usage = dataset.groupby("Sleep_Hours_Per_Night")["Avg_Daily_Usage_Hours"].mean().reset_index()

plt.figure(figsize=(8,6))
sns.lineplot(data=sleep_usage, x="Sleep_Hours_Per_Night", y="Avg_Daily_Usage_Hours", marker='o', linewidth=2)
plt.title("Average Daily Usage by Sleep Hours")
plt.xlabel("Sleep Hours Per Night")
plt.ylabel("Average Daily Usage (Hours)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

* Students sleeping less than 6 hours tend to have higher daily social media usage.

### Academic performance (Box plot)

```python
# Box plot

plt.figure(figsize=(6,6))
sns.boxplot(data=dataset, x="Affects_Academic_Performance", y="Avg_Daily_Usage_Hours", palette="Set2")
plt.title("Impact of Social Media on Academic Performance")
plt.xlabel("Affects Academic Performance (Yes/No)")
plt.ylabel("Average Daily Usage (Hours)")
plt.tight_layout()
plt.show()
```

* Students who say social media affects their academics have higher daily usage.

### Social Interaction (Bar Chart)

```python
# Bar chart
conflict_usage = dataset.groupby("Conflicts_Over_Social_Media")["Avg_Daily_Usage_Hours"].mean().reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=conflict_usage,
            x="Conflicts_Over_Social_Media",
            y="Avg_Daily_Usage_Hours",
            palette="pastel")
plt.title("Average Daily Usage by Number of Social Media Conflicts")
plt.xlabel("Number of Conflicts Over Social Media")
plt.ylabel("Average Daily Usage (Hours)")
plt.tight_layout()
plt.show()
```



*  Students with more than 3 social conflicts tend to spend more time on social media.


# Aggregation and Insights

## Use groupby and aggregation to find average addiction level across different:
* Genders
* Age groups
* Education levels

```python
# Gender
print("Average Addiction Score by Gender:")
print(dataset.groupby("Gender")["Addicted_Score"].mean())

# Age Groups
bins = [17, 19, 21, 23, 25]
labels = ["18-19", "20-21", "22-23", "24-25"]
dataset["Age_Group"] = pd.cut(dataset["Age"], bins=bins, labels=labels, right=True)
print("\nAverage Addiction Score by Age Group:")
print(dataset.groupby("Age_Group")["Addicted_Score"].mean())

# Academic Level
print("\nAverage Addiction Score by Academic Level:")
print(dataset.groupby("Academic_Level")["Addicted_Score"].mean())
```

# Functions, Loops an Conditionals

## Create custom functions to:
* Classify risk level (Low/Medium/High) based on usage hours
* Suggest digital detox strategies using if-else blocks

```python
# Risk level
dataset["Risk_Level"] = dataset["Avg_Daily_Usage_Hours"].apply(
    lambda h: "Low" if h < 3 else "Medium" if h <= 5 else "High"
)

# Detox strategy
dataset["Detox_Strategy"] = dataset["Risk_Level"].apply(
    lambda r: "Keep balanced" if r == "Low" else "Limit to 2 hrs/day" if r == "Medium" else "Weekend detox"
)

dataset[["Avg_Daily_Usage_Hours", "Risk_Level", "Detox_Strategy"]].head()
```
